# Model

```python
import matplotlib.pyplot as plt
from pytorch_forecasting import TemporalFusionTransformer, TimeSeriesDataSet
from pytorch_forecasting.metrics import QuantileLoss, MAE, MAPE, RMSE, MSE, R_Squared
from pytorch_forecasting.models.temporal_fusion_transformer.tuning import optimize_hyperparameters
import pytorch_lightning as pl
import torch
import numpy as np


# Define the maximum encoder and prediction lengths
max_encoder_length = 20
max_prediction_length = 4

# Create the TimeSeriesDataSet for training
training = TimeSeriesDataSet(
    train,
    time_idx="time_idx",
    target="Price",
    group_ids=["Series"],
    static_categoricals=["Market", "Length", "Grade_Type", "Width", "Height"],
    time_varying_known_categoricals=[],
    time_varying_known_reals=["time_idx"] + time_varying,
    time_varying_unknown_reals=["Price"] + unknown_future,
    max_encoder_length=max_encoder_length,
    max_prediction_length=max_prediction_length,
)

# Create validation and test datasets from the training dataset
validation = TimeSeriesDataSet.from_dataset(training, validation, predict=True, stop_randomization=True)
test = TimeSeriesDataSet.from_dataset(training, test, predict=True, stop_randomization=True)

# Step 2: Create DataLoaders
batch_size = 64
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size, num_workers=0)
test_dataloader = test.to_dataloader(train=False, batch_size=batch_size, num_workers=0)

# Step 3: Define the TFT Model
model = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=0.03,
    hidden_size=16,
    attention_head_size=1,
    dropout=0.1,
    hidden_continuous_size=8,
    output_size=7,  
    loss=QuantileLoss(),
    log_interval=10,
    reduce_on_plateau_patience=4,
)

# Step 4: Train the Model 
trainer = pl.Trainer(
    max_epochs=65,  
    gpus=1 if torch.cuda.is_available() else 0,  
    weights_summary="top",
    gradient_clip_val=0.1,
    callbacks=[pl.callbacks.EarlyStopping(monitor="val_loss", patience=5, min_delta=1e-4)],  
)

trainer.fit(
    model,
    train_dataloaders=train_dataloader,
    val_dataloaders=val_dataloader,
)

# Step 5: Optimize Hyperparameters using Optuna 
study = optimize_hyperparameters(
    train_dataloader,
    val_dataloader,
    model_path="optuna_test",
    n_trials=200,  
    max_epochs=50,
    gradient_clip_val_range=(0.01, 1.0),
    hidden_size_range=(8, 256),  
    hidden_continuous_size_range=(8, 256),  
    attention_head_size_range=(1, 16), 
    learning_rate_range=(0.001, 0.05),  
    dropout_range=(0.1, 0.5),  
    trainer_kwargs=dict(limit_train_batches=30, gpus=1 if torch.cuda.is_available() else 0),  
    reduce_on_plateau_patience=4,
    use_learning_rate_finder=False, 
)

# Step 6: Evaluate the Model
best_model_path = study.best_trial.user_attrs["model_path"]
best_model = TemporalFusionTransformer.load_from_checkpoint(best_model_path)

```

```python
# Function to calculate metrics
def calculate_metrics(dataloader, model):
    actuals = []
    predictions = []
    for batch in dataloader:
        x, y = batch
        output = model(x)
        actuals.append(y[0].cpu().numpy())
        predictions.append(output.median().cpu().numpy())

    actuals = np.concatenate(actuals)
    predictions = np.concatenate(predictions)

    mae = MAE()(torch.tensor(predictions), torch.tensor(actuals)).item()
    mape = MAPE()(torch.tensor(predictions), torch.tensor(actuals)).item()
    rmse = RMSE()(torch.tensor(predictions), torch.tensor(actuals)).item()
    mse = MSE()(torch.tensor(predictions), torch.tensor(actuals)).item()
    r_squared = R_Squared()(torch.tensor(predictions), torch.tensor(actuals)).item()

    return mae, mape, rmse, mse, r_squared

# Calculate metrics on train data
train_mae, train_mape, train_rmse, train_mse, train_r_squared = calculate_metrics(train_dataloader, best_model)

# Calculate metrics on test data
test_mae, test_mape, test_rmse, test_mse, test_r_squared = calculate_metrics(test_dataloader, best_model)

# Print metrics
print("Training Metrics:")
print(f"MAE: {train_mae}, MAPE: {train_mape}, RMSE: {train_rmse}, MSE: {train_mse}, R-squared: {train_r_squared}")
print("\nTest Metrics:")
print(f"MAE: {test_mae}, MAPE: {test_mape}, RMSE: {test_rmse}, MSE: {test_mse}, R-squared: {test_r_squared}")
```

```python
# Step 7: Visualize Results
actuals = torch.cat([y[0] for x, y in iter(test_dataloader)])
predictions = best_model.predict(test_dataloader)

plt.figure(figsize=(10, 6))
plt.plot(actuals, label="Actual")
plt.plot(predictions, label="Predicted")
plt.legend()
plt.show()
```
